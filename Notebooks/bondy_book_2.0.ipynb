Let's start with the columns with the least amount of nulls. What's going on here?

```python
# check out records where first serve points won percentage is null
df[df['player_1_first_serve_win_%'].isnull()]
```

Don't need matches where the player retired after 1 game in the entire match.

```python
# check out records where first serve percentage is null
df[df['player_1_first_serve_%'].isnull()]
```

This is definitely janky data we don't need. Service points for player 1 and player 2 definitely do not make sense, and we can definitely afford to drop these. No need to look this info up.

```python
# drop janky records
df = df[df['player_1_first_serve_%'].notnull()]
```

```python
# verify
df.shape
```

```python
# check out records where first serve percentage is null
df[df['player_2_first_serve_%'].isnull()]
```

Definitely drop those retired matches. And let's drop these other ones for speed, second serve stats are missing.

```python
# check out records where first serve percentage is null
df[df['player_2_first_serve_win_%'].isnull()]
```

These are all retired matches with the exceptions of the aforementioned matches. Drop.

```python
# drop, check
df = df[df['player_2_first_serve_%'].notnull()]
df.shape
```
